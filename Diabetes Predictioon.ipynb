```python
# apply labelencoding on required columns
# import labelencoder
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
 # take all columns in 1 list
cols = data.columns
cols
```

```python
#we drop "Age column" because it is already in numeric form
cols1 = cols.drop(['Age'])
cols1
```

```python
data1 = data[cols1]
data1
```

```python
data1.head()
```

```python
# apply labelencoding on data1
for columns in cols1:
    data1[columns]=le.fit_transform(data1[columns])
    

```

```python
#add Age column in data1
data1['Age']= data['Age']
data1520 rows × 17 columns
                   
```

```python
Y = pd.DataFrame()
Y['class']=data1['class']
```

```python
c = data1.columns
```

```python
c1= c.drop(['class'])
```

```python
X =data1[c1]
X
```

```python
X.head()
```

```python
# Split Data
from sklearn.model_selection import train_test_split
```

```python

xtrain, xtest, ytrain, ytest = train_test_split(X, Y, test_size=0.2)
```

```python
# now prepare model
from sklearn.linear_model import LogisticRegression
lr= LogisticRegression()
```

```python
lr.fit(xtrain,ytrain)
```

```python
lr.coef_
```

```python
pred = lr.predict(xtest)
pred
```

```python
pred = pd.DataFrame(pred)
pred.head()
```

```python
ytest.head()
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
ac = accuracy_score(ytest,pred)
print("accuracy_score :",ac)
```

```python
cr = classification_report(ytest, pred)
print("classification report is :\n",cr)
```

```python
cm = confusion_matrix(ytest,pred)
print("confusion matrix ic :\n",cm)
```
